```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data_collecting_action_frame.csv", dtype=object)
df.head()
```

```python
ndf = df[['frame_num', 'nose', 'left_eye', 'right_eye', 'left_ear', 'right_ear',
       'left_shoulder', 'right_shoulder', 'left_elbow', 'right_elbow',
       'left_wrist', 'right_wrist', 'left_hip', 'right_hip', 'left_knee',
       'right_knee', 'left_ankle', 'right_ankle','Xbound', 'Ybound', 'LE_angle', 'RE_angle', 'action']]
```

```python
import re
import math
frame_width = 1920
frame_height = 1080
def append_features(data_frame,idx):
    features_list = []
    for kpts in data_frame.keys()[1:-1]:
        features = data_frame.iloc[idx][kpts]
        # print(kpts)
        if (kpts=="LE_angle") or (kpts=="RE_angle"):
            features = np.round(float(features),2)
            features *= (math.pi/360)
        else:
            features = turn2list(features)
            if kpts == "Xbound":
                features[0]/=frame_width
                features[1]/=frame_width
            elif kpts == "Ybound":
                features[0]/=frame_height
                features[1]/=frame_height
            # features = np.round(features,2)
            else:
                features[0]/=frame_width
                features[1]/=frame_height

        features_list.append(features)
    return np.concatenate(features_list, axis=None,dtype=np.float64)


def turn2list(input_string):
    # Input string
    input_string = str(input_string)
    # Use regular expression to extract numbers
    input_string = input_string.split('[')[-1]
    input_string = input_string.split(']')[0]
    numbers = input_string.split(' ')
    # print(numbers)
    number_list = []
    for num in numbers:
        if num!='':
            if num.split(",")!=0:
                num = num.split(",")[0]
            number_list.append(float(num))
    # Convert strings to float and create a list
    return number_list
```

```python
from tqdm import tqdm 
df_dict = {}
df_time = []
df_features = []
df_action = []
for k in tqdm(range(len(ndf))):
    df_time.append(int(k))
    df_features.append(append_features(ndf,k))
    df_action.append(ndf.iloc[k].action)
df_dict["time"] = df_time
df_dict["features"] = df_features
df_dict["action"] = df_action
new_dataframe = pd.DataFrame(df_dict)
```
